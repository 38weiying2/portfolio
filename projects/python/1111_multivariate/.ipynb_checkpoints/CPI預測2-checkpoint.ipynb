## <ins>動機與目的</ins>

根據近期的新聞發現一件事，當CPI下降時，股市可能會上漲。  
由於s&p500是領先指標，CPI是落後指標，因此CPI受到股價影響是有可能的。

CPI是一種經濟的結果和現象，它不會直接影響股市，但是CPI會影響政策制定，而政策會影響股市。例如央行是否升息降息、是否調整存款準備金比率等。  
換句話說，CPI與股市(以標普500指數為例(S&P500))沒有直接關係，但 CPI 的變化與標普500指數的表現可能存在相關性。一般而言，當 CPI 上升時，可以表明經濟存在通膨。高通膨會導致消費者支出和投資減少，進而導致股價下跌。此外，當美聯儲加息以對抗通貨膨脹時，可能會導致股價下跌。另一方面，當CPI穩定或下降時，可以表明經濟中很少或沒有通脹。這對股價來說可能是一個積極信號，因為低通脹可以為公司帶來更有利的經濟環境，從而提高收益並使股票對投資者更具吸引力。

本益比是一種估值指標，顯示投資者願意為公司的收益支付多少。  
投資者願意在通貨膨脹率高時為一定水平的收益支付更少的費用，而在通貨膨脹率低（並且預計會保持這種水平）時為一定水平的收益支付更多費用。  
因此本益比與CPI也是息息相關的。

根據reference文章表示，經濟理論表明，通貨膨脹率隨著失業率下降而上升。因為根據菲利普斯曲線，較低的失業率意味著人們會花更多的錢，從而對價格造成更大的壓力，而通膨與失業率也可能呈正向關係(政策)。通膨與失業率可以說是互相影響，較高的失業率等同於較低的通貨膨脹率。當更多的人在工作時，他們就有能力消費，從而導致需求增加。價格（通貨膨脹）很快就會隨之而來。當失業率上升時，情況正好相反。  
因此也可以說是CPI影響失業率，而失業率又影響CPI。


因此本project目的是在研究當代表的股價、本益比與失業率的漲跌，是否會使CPI會有連帶的漲與跌、看能否預測CPI的時間序列。

### Stationarity Test

```python
from statsmodels.tsa.stattools import adfuller
def stationary_TS(ts):
    name = ts.name
    print(f"{ts.name} p-value: {adfuller(ts)[1]}")
    if adfuller(ts)[1] > 0.05:
        print(f'{ts.name} p-value > 0.05，我們不能拒絕原假設(也就是說我們有一個非平穩的數據集)。')
    else:
        print(f'{ts.name} p-value < 0.05，我們拒絕原假設(也就是說我們有一個平穩的數據集)。')
```

```python
for i in dataDF:
    stationary_TS(dataDF[i])
```

#### <span style='background:#F9FAB6'>大多數都是非平穩，因此畫出ACF圖來看如何調整成平穩的</span>

### ACF and PACF Plots

```python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
for col in dataDF.columns:
    fig, ax = plt.subplots(1,2, figsize=(20,4))
    plot_acf(dataDF[col], zero=True, lags=40, ax=ax[0], title=f'ACF - {col}')
    plot_pacf(dataDF[col], zero=True, method='ywm', lags=40, ax=ax[1], title=f'PACF - {col}')
    plt.show()
```

```python
from statsmodels.tsa.stattools import grangercausalitytests
granger1 = grangercausalitytests(diffDF[['CPIAUCSL_PC1', 'SP500']].dropna(), maxlag=12, verbose=False)
granger2 = grangercausalitytests(diffDF[['CPIAUCSL_PC1', 'UNRATE']].dropna(), maxlag=12, verbose=False)
granger3 = grangercausalitytests(diffDF[['CPIAUCSL_PC1', 'SP500_PE']].dropna(), maxlag=12, verbose=False)

print('SP500    ',[round(granger1[i+1][0]['ssr_chi2test'][1],4) for i in range(12)])
print('UNRATE   ',[round(granger2[i+1][0]['ssr_chi2test'][1],4) for i in range(12)])
print('SP500_PE ',[round(granger3[i+1][0]['ssr_chi2test'][1],4) for i in range(12)])
```

### Cointegration Test
(如果X的變化先於Y的變化，那麼X可能是Y的原因。)

```python
from statsmodels.tsa.stattools import coint

def coint_func(x,y):
    coint_pval = coint(x, y)[1]
    print("p-value: ", coint_pval)
    if coint_pval < 0.05:
        print("They are cointegrated (協整的)")
    else:
        print("They are not cointegrated (不協整的)")
```

```python
coint_func(diffDF['SP500'], diffDF['CPIAUCSL_PC1'])
coint_func(diffDF['UNRATE'], diffDF['CPIAUCSL_PC1'])
coint_func(diffDF['SP500_PE'], diffDF['CPIAUCSL_PC1'])
```

#### <span style='background:#F9FAB6'>SP500、UNRATE可能是影響cpi的原因，而SP500_PE則可能不是</span>

### <ins>2.VAR Modeling</ins>

```python
from statsmodels.tsa.api import VAR
train = diffDF[:90]
test = diffDF[90:]
print(f'Train: {len(train)}, Test: {len(test)}')
```

### Find VAR order p, The VAR will determine the best p values that minimize each of the four information criteria scores: AIC, BIC, FPE, and HQIC.

```python
model = VAR(endog=train, freq='MS')
result = model.select_order(maxlags=10)
result.summary()
```

```python
result.selected_orders
```

### Use the AIC score

#### <span style='background:#F9FAB6'>AIC為10，maxlags也設10</span>

```python
results = model.fit(maxlags=10, ic='aic')
```

```python
results.summary()
```

```python
results.resid_corr
```

### Forecasting

```python
lag_order = results.k_ar
# lag_order = 7
past_x = train[-lag_order:].values
n = test.shape[0]
forecast, lower, upper = results.forecast_interval(past_x, steps=n)

ax = train.iloc[:-lag_order, 1].plot(style='g', figsize=(20,4))
pred_forecast = pd.Series(forecast[:, 1], index=test.index).plot(style='r', ax=ax)
pred_lower = pd.Series(lower[:, 1], index=test.index).plot(style='k--', ax=ax)
pred_upper = pd.Series(upper[:, 1], index=test.index).plot(style='k--', ax=ax)
plt.fill_between(test.index, lower[:,1], upper[:,1], alpha=0.3, color='r')
plt.title('Forecasting cpi')
plt.legend(['Train','Forecast'])
plt.show()
```

### Rolling Forecast Using SARIMAX
SARIMAX 模型通過添加外生變量（輸入變量）的影響擴展了 SARIMA(p,d,q)(P,D,Q)m 模型。 因此，我們可以將現值𝑦𝑡簡單地表示為 SARIMA(p,d,q)(P,D,Q)m 模型

```python
from sklearn.metrics import mean_absolute_percentage_error as mape
from sklearn.metrics import r2_score
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

target_train = target[:90]
target_test = target[90:]
train_len = len(target_train)
total_len = target.shape[0]
WINDOW = 1
pred_SARIMAX = []
for i in range(train_len, total_len, WINDOW):
    model = SARIMAX(target[:i], featureData[:i], order=(1,0,1), seasonal_order=(2,0,0,4))
    res = model.fit(disp=False)
    predictions = res.get_prediction(exog=featureData)
    oos_pred = predictions.predicted_mean.iloc[-WINDOW:]
    pred_SARIMAX.extend(oos_pred)
```

```python
predDF = pd.DataFrame({'actual': target_test})
predDF['pred_SARIMAX'] = pred_SARIMAX
predDF.head(3).append(predDF.tail(3))
```

```python
import numpy as np

SARIMAX_mape_diff = np.round(mape(predDF.actual, predDF.pred_SARIMAX), 4)
SARIMAX_r2_diff = np.round(r2_score(predDF.actual, predDF.pred_SARIMAX), 4)
predDF.plot(figsize=(20,6), title = f'MAPE: {SARIMAX_mape_diff} and $R^2$ = {SARIMAX_r2_diff}')
plt.show()
```

#### <span style='background:#F9FAB6'>可以看到SARIMAX效果還不錯，是相當合適的模型</span>

```python
from prophet import Prophet
from prophet.utilities import regressor_coefficients
import seaborn as sns

def prophet_modeling(seasonality_mode, train, test, dataProphet, FREQ="MS"):
    m = Prophet(daily_seasonality = True, seasonality_mode=seasonality_mode)
    m.add_regressor('sp500')
    m.add_regressor('unrate')
    m.fit(train)
    # Forecast Future
    future = m.make_future_dataframe(periods=test.shape[0], freq = FREQ, include_history=True)
    future['sp500'] = dataProphet['sp500']
    future['unrate'] = dataProphet['unrate']

    # prediction = m.predict(future.dropna()
    prediction = m.predict(future)
    
    # Plot all
    m.plot(prediction, figsize=(20, 4))
    plt.title(f"Prediction of the Future CPI for the next {test.shape[0]} steps ({FREQ})")
    plt.xlabel("Date")
    plt.ylabel("CPI")
    plt.show()
    
    # plot on test data 
    prediction['y'] = dataProphet['y']
    prediction['Date'] = dataProphet['ds']
    prediction= prediction.set_index('Date')
    
    fig, ax = plt.subplots(figsize=(20,4))
    testDF = prediction[['y', 'yhat_lower', 'yhat_upper', 'yhat']][90:]
    Prophet_mape = np.round(mape(testDF.y, testDF.yhat), 4)
    Prophet_r2 = np.round(r2_score(testDF.y, testDF.yhat), 4)
    
    ax = sns.lineplot(data=testDF)
    ax.fill_between(testDF.index, testDF.yhat_lower, testDF.yhat_upper, alpha=0.3)
    ax.set(title= f' CPI Forecase - actual vs. predicted; Test MAPE: {Prophet_mape}; '+\
           f'Test $R^2$: {Prophet_r2}', xlabel='Date', ylabel='CPI')
    plt.show()
    # side_by_side('SARIMAX MAPE: ', SARIMAX_mape, 'Prophet MAPE: ', Prophet_mape)
    # print(regressor_coefficients(m))
    return m, Prophet_mape, Prophet_r2
```

### dataProphet_diff

### <ins>5.Machine Learning - Linear and Nonlinear Regression Models (機器學習 - 線性和非線性回歸模型)</ins>

### Linear Regression Models (線性回歸模型)

```python
from sklearn.preprocessing import PolynomialFeatures, StandardScaler, normalize
from sklearn.linear_model import Lasso, Ridge, LinearRegression, SGDRegressor, ARDRegression, Perceptron
from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
import statsmodels.api as sm
import numpy as np
```

```python
from datetime import datetime
skLR = LinearRegression(fit_intercept=True)
polyReg = LinearRegression(fit_intercept=True) #normalize=True,
ridge = Ridge(alpha=1, solver='sag', random_state=42)
lasso = Lasso(alpha=0.01, random_state=42)
SGD = SGDRegressor(max_iter=5000, eta0=0.00001, power_t=0.15, shuffle = False) 
ARD = ARDRegression()
svr_linear = SVR(kernel="linear", C=1000)
# Nonlinear Regressor
regrNN = MLPRegressor(hidden_layer_sizes=(200, 100, 50, 25), activation = 'relu', shuffle = False,
                      alpha=0.0001, random_state=1, max_iter=50000)
svr_nonlinear = SVR(kernel="rbf", C=1000)
```

```python
def train_test_process(listModels, listModelNames, trainData, testData, trainTarget, testTarget):
    print("模型訓練與測試\n")
    modelOLS = sm.OLS(trainTarget, trainData) # The benchmark model
    OLS_fit = modelOLS.fit()
    summary = OLS_fit.summary2()
    print("statsmodels OLS R^2 (Training Data Only) = ", summary.tables[0][0][6], summary.tables[0][1][6])
    prediction = OLS_fit.predict(testData)
    print("statsmodels OLS R^2 (Testing Data Only) = ", r2_score(testTarget, prediction))

    nameList = ['Ordinary Least Squares (OLS)']
    scoreList_Train=[float(summary.tables[0][1][6])]
    scoreList_Test=[r2_score(testTarget, prediction)]
    for aModel, aModelName in zip(listModels, listModelNames):
        process_starttime = datetime.now()
        if aModelName == 'Polynomial Regressor':
            poly = PolynomialFeatures(degree=2, include_bias=False)
            poly_features_train = poly.fit_transform(trainData)
            poly_features_test = poly.fit_transform(testData)
            aModel.fit(poly_features_train, trainTarget)
            y_hat = aModel.predict(poly_features_test)
            print ("\n..... "+aModelName+" 模型訓練花費時間 %s" % (datetime.now() - process_starttime))
            score_r2_train = aModel.score(poly_features_train, trainTarget).round(4) 
            score_r2_test = aModel.score(poly_features_test,  testTarget).round(4)
        else:            
            aModel.fit(trainData, trainTarget)
            y_hat = aModel.predict(testData)
            print ("\n..... "+aModelName+" 模型訓練花費時間 %s" % (datetime.now() - process_starttime))
            score_r2_train = aModel.score(trainData, trainTarget).round(4)
            score_r2_test = aModel.score(testData,  testTarget).round(4)
        nameList.append(aModelName)
        scoreList_Train.append(score_r2_train)
        scoreList_Test.append(score_r2_test)
    resultDF = pd.DataFrame(data = list(zip(nameList, scoreList_Train, scoreList_Test)),
                            columns=['Name of Regressor', 'Training r2', 'Testing r2'])
    resultDF.sort_values("Training r2", axis = 0, ascending = False, inplace = True)
    resultDF.reset_index(drop=True,  inplace = True)
    resultDF['Diff%'] = np.round((resultDF['Training r2']-resultDF['Testing r2'])/resultDF['Training r2'],
                                 4)*100
    return resultDF
```

```python
listModelNames = ['Linear Regressor', 'Polynomial Regressor', 'Ridge Regressor', 'Lasso Regressor', 
                  'SGD', ' Bayesian ARD regression', 'Support Vector Linear Regressor',
                  'Support Vector Nonlinear Regressor', 'Neural Nework Regressor']
listModels = [skLR, polyReg, ridge, lasso, SGD, ARD, svr_linear, svr_nonlinear, regrNN]
```

### diff

```python
target_train = target[:90]
feature_train = featureData[:90]
target_test = target[90:]
feature_test = featureData[90:]
resultDF = train_test_process(listModels, listModelNames, feature_train, feature_test, 
                              target_train, target_test)
resultDF
```
